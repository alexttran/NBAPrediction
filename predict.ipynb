```python
import pandas as pd
```

```python
df = pd.read_csv("nba_games.csv", index_col=0)
```

```python
df = df.sort_values("date")
```

```python
df = df.reset_index(drop=True)
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
def add_target(team):
    team["target"] = team["won"].shift(-1)
    return team

df = df.groupby("team", group_keys=False).apply(add_target)
```

```python
df["target"][pd.isnull(df["target"])] = 2
df["target"] = df["target"].astype(int, errors="ignore")
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split)
```

```python
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]
```

```python
selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
sfs.fit(df[selected_columns], df["target"])
```

```python
predictors = list(selected_columns[sfs.get_support()])
```

```python
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []

    seasons = sorted(data["season"].unique())

    for i in range(start, len(seasons), step):
        season = seasons[i]

        train = data[data["season"] < season]
        test = data[data["season"] == season]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)

        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score

predictions = predictions[predictions["actual"] != 2]
accuracy_score(predictions["actual"], predictions["prediction"])
```

```python
def find_team_averages(team):
    team = team.sort_values("date")
    numeric = team.select_dtypes(include="number")
    rolling = numeric.rolling(10).mean()
    # Re-attach non-numeric columns if needed
    for col in team.columns:
        if col not in numeric.columns:
            rolling[col] = team[col].values
    return rolling

df_rolling = df_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages)

#def find_team_averages(team):
#    rolling = team.rolling(10).mean()
#    return rolling

#df_rolling = df_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages)
```

```python
rolling_cols = [f"{col}_10" for col in df_rolling.columns]
df_rolling.columns = rolling_cols

df = pd.concat([df, df_rolling], axis=1)
```

```python
df = df.dropna()
```

```python
def shift_col(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col

def add_col(df, col_name):
    return df.groupby("team", group_keys=False).apply(lambda x: shift_col(x, col_name))

df["home_next"] = add_col(df, "home")
df["team_opp_next"] = add_col(df, "team_opp")
df["date_next"] = add_col(df, "date")
```

```python
def predict_winner(team1, team2, model, full_df, predictors):
    # Find the most recent matchup between these two teams
    matchup = full_df[(full_df["team_x"] == team1) & (full_df["team_y"] == team2)]
    if matchup.empty:
        raise ValueError(f"No matchup found between {team1} and {team2} in the data.")
    # Use the latest available matchup
    matchup = matchup.sort_values("date_next").iloc[-1]
    X = matchup[predictors].values.reshape(1, -1)
    pred = model.predict(X)[0]
    return team1 if pred == 1 else team2
```

```python
winner = predict_winner("SAS", "POR", rr, full, predictors)
print("Predicted winner:", winner)
winner = predict_winner("POR", "SAS", rr, full, predictors)
print("Predicted winner:", winner)
```
